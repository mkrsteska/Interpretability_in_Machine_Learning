```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
```

```python
# read the dataset
df = pd.read_csv('../../data/raw_data/Breast_Cancer.csv')
```

```python
# show the first 5 rows
df.head()
```

```python
# dataset dimension
df.shape
```

```python
# attributes
df.columns
```

```python
# show the column types
df.dtypes
```

```python
# check if there are NaN values
df.isna().values.any()
```

```python
# number of missing values per feature
df.isnull().sum().sort_values(ascending=False)
```

```python
# drop the columns 'id' and 'Unnamed'
df = df.drop(columns=['id','Unnamed: 32'])
```

```python
# select only the numerical features
df_numeric = df.select_dtypes(include=np.number)

# plot the distribution of each numerical feature
fig = plt.figure(figsize=(16, 12))
fig.subplots_adjust(hspace=0.4, wspace=0.2)
for i, col in enumerate(df_numeric.columns):
    ax = fig.add_subplot(6, 5, i+1)
    hist = sns.distplot(df_numeric[col], ax=ax)
```

```python
# write dataframe to csv file
df.to_csv('../../data/clean_data/Breast_Cancer.csv')
```
